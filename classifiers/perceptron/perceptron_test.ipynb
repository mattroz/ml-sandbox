```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from one_layer_perceptron import Perceptron
```

```python
# Generate synthetic data
x1 = 1 + .5 * np.random.randn(500)
x2 = 4.5 + .5 * np.random.randn(500)
y1 = 7 + 2 * np.random.randn(500)
y2 = 5 + 15 * np.random.rand(500)
plt.scatter(x1,y1,color='red')
plt.scatter(x2,y2,color='blue')
plt.show()
```

```python
# Preprocess generated data
trainset = np.concatenate((x1, x2))
y = np.concatenate((y1, y2))
trainset = np.array(list(zip(trainset, y)))
target = np.array((-np.ones(500), np.ones(500))).ravel()
```

```python
# Train perceptron
perceptron = Perceptron(n_iters=240)
perceptron.fit(trainset, target)
```

```python
# Plot perceptron error progress through epochs
plt.plot(range(1, len(perceptron.errors) + 1), perceptron.errors, marker='o')
plt.xlabel('Epochs')
plt.ylabel('Num of missclassif-s')
plt.show()
```
